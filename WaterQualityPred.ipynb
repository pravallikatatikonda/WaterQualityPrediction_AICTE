```python
pollutants=['O2', 'NO3', 'NO2', 'SO4', 'PO4', 'CL']
```

### Week 2-23 June **2025**

```python
initial_shape = df.shape
df = df.dropna(subset=pollutants)
print(f"Dropped {initial_shape[0] - df.shape[0]} rows with missing pollutant values.\n")
```

```python
# Features (X): Station ID and Year
# Target (y): Pollutant levels
X = df[['id', 'year']]
y = df[pollutants]
```

```python
# Convert categorical 'id' column into numerical one-hot encoded columns
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
model = MultiOutputRegressor(
    RandomForestRegressor(n_estimators=100, random_state=42)
)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f"{pollutant}:")
    print("  MSE:", mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print("  R2 :", r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
def predict_pollutants(station_id, year_input, model, model_columns):
    input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
    input_encoded = pd.get_dummies(input_data, columns=['id'])

    # Add any missing one-hot encoded columns
    for col in model_columns:
        if col not in input_encoded:
            input_encoded[col] = 0

    # Ensure column order matches training data
    input_encoded = input_encoded[model_columns]

    prediction = model.predict(input_encoded)[0]
    return prediction
```

```python
station_id = '22'
year_input = 2024

predicted_pollutants = predict_pollutants(station_id, year_input, model, X_encoded.columns.tolist())

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```
